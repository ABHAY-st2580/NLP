## Removing URLS

```python
def remove_urls(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

## Removing Punctuations

```python
import string 
```

```python
punc = string.punctuation
```

```python
punc
```

```python
def removing_punct(text):
    for char in punc:
        text = text.replace(char, '')
    return text
```

```python
df['text'][3]
```

```python
removing_punct(df['text'][3])
```

```python
def rem_punct(text):
    return text.translate(str.maketrans('', '', punc)) ## Much faster than above...
```

```python
df['text'][3]
```

```python
rem_punct(df['text'][3])
```

## Chat Word Treatement

```python
chat_slang_dict = {
    "u": "you",
    "ur": "your",
    "r": "are",
    "y": "why",
    "b4": "before",
    "gr8": "great",
    "l8r": "later",
    "btw": "by the way",
    "idk": "i don't know",
    "imo": "in my opinion",
    "imho": "in my humble opinion",
    "lol": "laughing out loud",
    "rofl": "rolling on the floor laughing",
    "lmao": "laughing my ass off",
    "omg": "oh my god",
    "ttyl": "talk to you later",
    "brb": "be right back",
    "afk": "away from keyboard",
    "np": "no problem",
    "thx": "thanks",
    "ty": "thank you",
    "yw": "you're welcome",
    "pls": "please",
    "plz": "please",
    "bc": "because",
    "cuz": "because",
    "coz": "because",
    "smh": "shaking my head",
    "tbh": "to be honest",
    "fyi": "for your information",
    "irl": "in real life",
    "jk": "just kidding",
    "dm": "direct message",
    "msg": "message",
    "wtf": "what the f***",
    "wth": "what the hell",
    "asap": "as soon as possible",
    "nvm": "never mind",
    "bff": "best friends forever",
    "ily": "i love you",
    "xoxo": "hugs and kisses",
    "gg": "good game",
    "noob": "beginner",
    "op": "overpowered",
    "idc": "i don't care",
    "ikr": "i know right",
    "wyd": "what are you doing",
    "wbu": "what about you",
    "hbu": "how about you",
    "gm": "good morning",
    "gn": "good night"
}
```

```python
def chat_word_treatment(text):
    new_text = []
    for w in text.split():
        if w.lower() in chat_slang_dict:
            new_text.append(chat_slang_dict[w.lower()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
text = 'imho he is the best'
```

```python
chat_word_treatment(text)
```

## Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def removing_stop_words(text):
    next_word = []
    for word in text.split():
        if word in stopwords.words('english'):
            next_word.append('')
        else:
            next_word.append(word)
    x = next_word[:]
    next_word.clear()
    return " ".join(x)
```

```python
df['text'][3].lower()
```

```python
removing_stop_words(df['text'][3].lower())
```

## Handling Emoji's

```python
## Handling the emojis by removing the emojis from the text..

import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons 😀
        "\U0001F300-\U0001F5FF"  # symbols & pictographs 🌍
        "\U0001F680-\U0001F6FF"  # transport 🚗
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # geometric shapes
        "\U0001F800-\U0001F8FF"  # supplemental arrows
        "\U0001F900-\U0001F9FF"  # supplemental symbols 🤖
        "\U0001FA00-\U0001FA6F"  # chess symbols
        "\U0001FA70-\U0001FAFF"  # extended symbols
        "\U00002702-\U000027B0"  # dingbats ✂️
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
text = "Hurry!! Your account is blocked 😡🔥 Click here 👉 http://fake.com"
clean = remove_emojis(text)

print(clean)
```
